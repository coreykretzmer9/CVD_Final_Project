```python
demographics_file = "../Final Project/ACS_16_5YR_B02001_with_ann.csv"
demographics_df = pd.read_csv(demographics_file, encoding='cp1252')
```

```python
demographics_df = demographics_df[demographics_df.columns.drop(list(demographics_df.filter(regex='Margin of Error')))]
demographics_df = demographics_df.drop(["Id", "Geography","Estimate; Total: - Two or more races: - Two races including Some other race", "Estimate; Total: - Two or more races: - Two races excluding Some other race, and three or more races"], axis=1)
demographics_df.rename(columns={'Id2': 'FIPS Code', 'Estimate; Total:': 'Population Total', "Estimate; Total: - White alone" : "White", "Estimate; Total: - Black or African American alone" : "Black", "Estimate; Total: - American Indian and Alaska Native alone" : "Native American", "Estimate; Total: - Asian alone" : "Asian", "Estimate; Total: - Native Hawaiian and Other Pacific Islander alone" : "Pacific Islander", "Estimate; Total: - Some other race alone" : "Other","Estimate; Total: - Two or more races:" : "Two or More Races"}, inplace=True)
demographics_df.head()
```

```python
demographics_df['White %'] = demographics_df.apply(lambda row: row['White'] / row['Population Total'], axis=1)
demographics_df['Black %'] = demographics_df.apply(lambda row: row['Black'] / row['Population Total'], axis=1)
demographics_df['Native American %'] = demographics_df.apply(lambda row: row['Native American'] / row['Population Total'], axis=1)
demographics_df['Asian %'] = demographics_df.apply(lambda row: row['Asian'] / row['Population Total'], axis=1)
demographics_df['Pacific Islander %'] = demographics_df.apply(lambda row: row['Pacific Islander'] / row['Population Total'], axis=1)
demographics_df['Other %'] = demographics_df.apply(lambda row: row['Other'] / row['Population Total'], axis=1)
demographics_df['Two or More Races %'] = demographics_df.apply(lambda row: row['Two or More Races'] / row['Population Total'], axis=1)
```

```python
demographics_df.head()
```

```python
demographics_df.columns
demographics_df = demographics_df[['FIPS Code', 'Population Total', 'White %', 'Black %', 'Native American %', 'Asian %', 'Pacific Islander %', 'Other %', 'Two or More Races %']]
demographics_df.head()
```

```python
diabetes_df.head(10)
```

```python
# Our original data-set has crude-rates based off of an adjusted population size of 100,000.  This shouldn't be an issue, we will just bin everything accordingly, relation will still hold.
diabetes_df = diabetes_df[["FIPS_code", "State", "County", "Age", "Estimate_type", "Estimate"]]
diabetes_df.head()
```

```python
diabetes_df = diabetes_df[diabetes_df["Age"] == "Crude"]
diabetes_df = diabetes_df[diabetes_df["Estimate_type"] == "Rate per 1,000"]
diabetes_df = diabetes_df.reset_index(drop=True)
diabetes_df.head()
```

```python
diabetes_df = diabetes_df.rename(columns={"Estimate" : "Estimated Rate Per 1,000"})
diabetes_df.head()
```

```python
diabetes_df = diabetes_df[['FIPS_code', 'State', 'County', 'Estimated Rate Per 1,000']]
diabetes_df.head()
```

```python
smokers_df.head(10)
```

```python
smokers_df = smokers_df[smokers_df.sex=="Both"].reset_index(drop=True)
smokers_df.head()
```

```python
smokers_df = smokers_df[['state', 'county', 'year', 'daily_mean']]
smokers_df.head()
```

```python
# Since our other data only includes years 2012-2016, deleting every row containing data for years besides 2012.
smokers_df = smokers_df[smokers_df.year==2012].reset_index(drop=True)
smokers_df.head()
```

```python
smokers_df = smokers_df.rename(columns={"daily_mean" : "% smoke daily"})
smokers_df.head()
```

```python
# Merge smokers with diabetes, look at shape to see if there's a discrepancy in row # between the two, if there is, we'll resolve this.
print(smokers_df.shape)
print(diabetes_df.shape)
# this shows us there are at least 35 rows that will not wholy match up.
```

```python
# Merge smokers with diabetes to get FIPS code, since we don't have corresponding FIPS code in smokers file.  Diabetes file is 
# formatted similarly in that it has both state "full name" and county name.
smokers_diabetes_df = pd.merge(diabetes_df, smokers_df, how='left', left_on=[diabetes_df.State, 'County'], right_on=[smokers_df.state, 'county'])
smokers_diabetes_df.head()
smokers_diabetes_df.to_csv("smokers_diabetes_combined_csv.csv")
```

```python
#smokers_diabetes_df.head()
print(smokers_diabetes_df.shape)
```

```python
smokers_diabetes_df = smokers_diabetes_df.dropna()
print(smokers_diabetes_df.shape)
```

```python
smokers_diabetes_df.head()
```

```python
smokers_diabetes_df = smokers_diabetes_df[['FIPS_code', 'State', 'County', 'Estimated Rate Per 1,000', '% smoke daily']]
smokers_diabetes_df = smokers_diabetes_df.rename(columns={'Estimated Rate Per 1,000' : 'Estimated Diabetes Rate Per 1,000', '% smoke daily' : '% Daily Smokers'})
```

```python
smokers_diabetes_df.head()
```

```python
fatty_df.head(10)
```

```python
# Remove unnecessary rows, everything compared will be as crude-percentages against a normalized population size, so, we can remove a shit-load of these boyos
fatty_df = fatty_df[(fatty_df.Age=='Crude') & (fatty_df.Sex=='Total') & (fatty_df.Estimate_type=='Percent')].reset_index(drop=True)
fatty_df.head()
```

```python
# Rename estimate to '% Obese', Remove unnecessary columns
fatty_df = fatty_df.rename(columns={'Estimate' : '% Population Obese'})
```

```python
fatty_df = fatty_df[['State', 'County', '% Population Obese']]
fatty_df.head()
```

```python
# Merge obese table with diabetes_smokers table
smokers_diabetes_obesity_df = pd.merge(smokers_diabetes_df, fatty_df, how='left', left_on=[smokers_diabetes_df.State, 'County'], right_on=[fatty_df.State, 'County'])
smokers_diabetes_obesity_df.head()
```

```python
smokers_diabetes_obesity_df = smokers_diabetes_obesity_df[['FIPS_code', 'State_x', 'County', 'Estimated Diabetes Rate Per 1,000', '% Daily Smokers', '% Population Obese']]
smokers_diabetes_obesity_df = smokers_diabetes_obesity_df.rename(columns={'State_x' : 'State', 'FIPS_code' : 'FIPS Code'})
```

```python
smokers_diabetes_obesity_df = smokers_diabetes_obesity_df.dropna().reset_index(drop=True)
smokers_diabetes_obesity_df.head()
```

```python
# Now, merge smokers-diabetus-fatty table with race table
print(demographics_df.shape)
print(smokers_diabetes_obesity_df.shape)
print(first_data_file_df.shape)
```

```python
smokers_diabetes_obesity_race_df = demographics_df.merge(smokers_diabetes_obesity_df, how="inner", on="FIPS Code")
smokers_diabetes_obesity_race_df.head()
```

```python
# Need to change the % of Race to *100, so that it is in-line with diabetes, obesity and smoking rate.
smokers_diabetes_obesity_race_df['% White'] = smokers_diabetes_obesity_race_df.apply(lambda row: row['White %'] * 100, axis=1)
smokers_diabetes_obesity_race_df['% Black'] = smokers_diabetes_obesity_race_df.apply(lambda row: row['Black %'] * 100, axis=1)
smokers_diabetes_obesity_race_df['% Native American'] = smokers_diabetes_obesity_race_df.apply(lambda row: row['Native American %'] * 100, axis=1)
smokers_diabetes_obesity_race_df['% Asian'] = smokers_diabetes_obesity_race_df.apply(lambda row: row['Asian %'] * 100, axis=1)
smokers_diabetes_obesity_race_df['% Pacific Islander'] = smokers_diabetes_obesity_race_df.apply(lambda row: row['Pacific Islander %'] * 100, axis=1)
smokers_diabetes_obesity_race_df['% Other'] = smokers_diabetes_obesity_race_df.apply(lambda row: row['Other %'] * 100, axis=1)
smokers_diabetes_obesity_race_df['% Two or More Races'] = smokers_diabetes_obesity_race_df.apply(lambda row: row['Two or More Races %'] * 100, axis=1)
smokers_diabetes_obesity_race_df.head()
```

```python
# smokers_diabetes_obesity_race_df.columns
smokers_diabetes_obesity_race_df = smokers_diabetes_obesity_race_df[['FIPS Code','State', 'County', 'Population Total','% White', '% Black', '% Native American',
       '% Asian', '% Pacific Islander', '% Other', '% Two or More Races', 
       'Estimated Diabetes Rate Per 1,000', '% Daily Smokers',
       '% Population Obese']].dropna().reset_index(drop=True)
smokers_diabetes_obesity_race_df.head()
```
